### Exercício 1

Crie um programa que classifique uma pessoa em uma das seguintes categorias com base na idade:

*   Criança (0-12 anos)
*   Adolescente (13 a 17 anos)
*   Adulto (18-64 anos)
*   Idosos (65 anos ou mais)  

```python
age = 12
print(age)
```

```python
def classificar_idade(age):
    if age >= 0 and age <= 13:
        return "criança"
    elif age >= 14 and age <= 17:
        return "adolescente"
    elif age >= 18 and age <= 64:
        return "adulto"
    elif age >= 65:
        return "idoso"
    else:
        return "idade inválida"

age = 12
categoria = classificar_idade(age)

print(f"A pessoa com {age} anos é classificada como: {categoria}")
```

### Exercício 3

Crie um programa que determine o tipo de triângulo (equilátero, isósceles ou escaleno) com base no comprimento de seus lados.

```python
def determinar_tipo_triangulo(side_1, side_2, side_3):
    if side_1 == side_2 == side_3:
        return "Equilátero"
    elif side_1 == side_2 or side_1 == side_3 or side_2 == side_3:
        return "Isósceles"
    else:
        return "Escaleno"

side_1 = 12
side_2 = 2
side_3 = 6
tipo_triangulo = determinar_tipo_triangulo(side_1, side_2, side_3)
print(f"O triângulo com lados {side_1}, {side_2} e {side_3} é do tipo: {tipo_triangulo}")
```

Esse código verifica se os lados do triângulo são todos iguais (Equilátero), dois lados iguais (Isósceles) ou todos os lados diferentes (Escaleno). Alterando os valores das variáveis side_1, side_2 e side_3 da testar diferentes comprimentos de lados.

```python
word = "pneumoultramicroscopicossilicovulcanoconiótico"
```

```python
def contar_vogais(palavra):
    vogais = "aeiouAEIOU"
    contador = 0
    for caractere in palavra:
        if caractere in vogais:
            contador += 1
    return contador

word = "pneumoultramicroscopicossilicovulcanoconiótico"
numero_vogais = contar_vogais(word)
print(f"A palavra '{word}' contém {numero_vogais} vogais.")
#pode alterar a variável word para qualquer outra palavra para testar diferentes cenários.
```

### Exercício 7

Projete um programa que simule as operações de um caixa eletrônico. Permite ao usuário fazer depósitos, saques e consultar saldo. Use loops `while` para manter a interação até que o usuário decida sair.

```python
def mostrar_menu():
    print("\n--- Caixa Eletrônico ---")
    print("1. Consultar saldo")
    print("2. Fazer saque")
    print("3. Fazer depósito")
    print("4. Sair")

def consultar_saldo(saldo):
    print(f"Seu saldo atual é: ${saldo:.2f}")

def fazer_saque(saldo):
    valor = float(input("Digite o valor do saque: $"))
    if 0 < valor <= saldo:
        saldo -= valor
        print(f"Saque de ${valor:.2f} realizado com sucesso!")
    else:
        print("Saldo insuficiente ou valor de saque inválido.")
    return saldo

def fazer_deposito(saldo):
    valor = float(input("Digite o valor do depósito: $"))
    if valor > 0:
        saldo += valor
        print(f"Depósito de ${valor:.2f} realizado com sucesso!")
    else:
        print("Valor de depósito inválido.")
    return saldo

def caixa_eletronico(saldo):
    while True:
        mostrar_menu()
        opcao = input("Escolha uma opção: ")
        if opcao == '1':
            consultar_saldo(saldo)
        elif opcao == '2':
            saldo = fazer_saque(saldo)
        elif opcao == '3':
            saldo = fazer_deposito(saldo)
        elif opcao == '4':
            print("Obrigado por usar o caixa eletrônico!")
            break
        else:
            print("Opção inválida, por favor tente novamente.")

# Executa o programa do caixa eletrônico com saldo inicial
cash = 3000
caixa_eletronico(cash)
```

A função mostrar_menu() exibe o menu de opções;

A função consultar_saldo() mostra o saldo atual;

A função fazer_saque() permite que o usuario faça o saque;

A função fazer_deposito() permite que o usuario faça o depósito.

A função caixa_eletronico() mantém o loop na interação até que o usuário decida sair.
