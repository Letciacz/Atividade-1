### Exerccio 1

Crie um programa que classifique uma pessoa em uma das seguintes categorias com base na idade:

*   Criana (0-12 anos)
*   Adolescente (13 a 17 anos)
*   Adulto (18-64 anos)
*   Idosos (65 anos ou mais)  

```python
age = 12
print(age)
```

```python
def classificar_idade(age):
    if age >= 0 and age <= 13:
        return "criana"
    elif age >= 14 and age <= 17:
        return "adolescente"
    elif age >= 18 and age <= 64:
        return "adulto"
    elif age >= 65:
        return "idoso"
    else:
        return "idade invlida"

age = 12
categoria = classificar_idade(age)

print(f"A pessoa com {age} anos  classificada como: {categoria}")
```

### Exerccio 3

Crie um programa que determine o tipo de tringulo (equiltero, issceles ou escaleno) com base no comprimento de seus lados.

```python
def determinar_tipo_triangulo(side_1, side_2, side_3):
    if side_1 == side_2 == side_3:
        return "Equiltero"
    elif side_1 == side_2 or side_1 == side_3 or side_2 == side_3:
        return "Issceles"
    else:
        return "Escaleno"

side_1 = 12
side_2 = 2
side_3 = 6
tipo_triangulo = determinar_tipo_triangulo(side_1, side_2, side_3)
print(f"O tringulo com lados {side_1}, {side_2} e {side_3}  do tipo: {tipo_triangulo}")
```

Esse cdigo verifica se os lados do tringulo so todos iguais (Equiltero), dois lados iguais (Issceles) ou todos os lados diferentes (Escaleno). Alterando os valores das variveis side_1, side_2 e side_3 da testar diferentes comprimentos de lados.

### Exerccio 7

Projete um programa que simule as operaes de um caixa eletrnico. Permite ao usurio fazer depsitos, saques e consultar saldo. Use loops `while` para manter a interao at que o usurio decida sair.

```python
def mostrar_menu():
    print("\n--- Caixa Eletrnico ---")
    print("1. Consultar saldo")
    print("2. Fazer saque")
    print("3. Fazer depsito")
    print("4. Sair")

def consultar_saldo(saldo):
    print(f"Seu saldo atual : ${saldo:.2f}")

def fazer_saque(saldo):
    valor = float(input("Digite o valor do saque: $"))
    if 0 < valor <= saldo:
        saldo -= valor
        print(f"Saque de ${valor:.2f} realizado com sucesso!")
    else:
        print("Saldo insuficiente ou valor de saque invlido.")
    return saldo

def fazer_deposito(saldo):
    valor = float(input("Digite o valor do depsito: $"))
    if valor > 0:
        saldo += valor
        print(f"Depsito de ${valor:.2f} realizado com sucesso!")
    else:
        print("Valor de depsito invlido.")
    return saldo

def caixa_eletronico(saldo):
    while True:
        mostrar_menu()
        opcao = input("Escolha uma opo: ")
        if opcao == '1':
            consultar_saldo(saldo)
        elif opcao == '2':
            saldo = fazer_saque(saldo)
        elif opcao == '3':
            saldo = fazer_deposito(saldo)
        elif opcao == '4':
            print("Obrigado por usar o caixa eletrnico!")
            break
        else:
            print("Opo invlida, por favor tente novamente.")

# Executa o programa do caixa eletrnico com saldo inicial
cash = 3000
caixa_eletronico(cash)
```

A funo mostrar_menu() exibe o menu de opes;

A funo consultar_saldo() mostra o saldo atual;

A funo fazer_saque() permite que o usuario faa o saque;

A funo fazer_deposito() permite que o usuario faa o depsito.

A funo caixa_eletronico() mantm o loop na interao at que o usurio decida sair.
